```python
"""
把excel数据转化为csv数据
"""
from data_process import *
import pandas as pd
# file_path = 'Data\Day_Population.xlsx'
# excel_to_csv(file_path)
data = pd.read_csv('Data\Day_Population.csv')
```

```python
from matplotlib import pyplot as plt
import numpy as np
plt.style.use('fivethirtyeight')
# 解决中文显示问题
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
# 生成figure对象
fig = plt.figure()
# 生成axes对象
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

y_quezhen = np.array(data['本土感染'].tolist())
y_wuzhengzhuang = np.array(data['本土无症状'].tolist())
y_all = np.array(data['合计'].tolist())
# 绘制散点
axes.plot(y_all, c="green", label=r'合计人数', ls='-', alpha=0.6, lw=2, zorder=2)
axes.plot(y_quezhen, c="red", label=r'本土确诊人数', ls='-', alpha=0.6, lw=2, zorder=2)
axes.plot(y_wuzhengzhuang, c="black", label=r'本土无症状', ls='-', alpha=0.6, lw=2, zorder=2)

# 设置图像标题
axes.set_title("每日增长人数曲线图")

axes.legend()
# 显示图像
plt.show()
```
